```python
def stock_market(url):
  for i in (equ['SYMBOL'][1:5]):
    url = f'https://www.screener.in/company/{i}/consolidated/'
    url_ = requests.get(url)
    soup = bs(url_.text)
    # print(soup)
    for j in soup.find_all('div', class_ = 'flex flex-column'):
      print(j.text)

    list_ = []
    for k in soup.find_all('span', class_ = 'number'):
      list_.append(k)

    data = {'name':['Market Cap','Current Price','High / Low','Stock P/E','Book Value','Dividend Yield','ROCE','ROE','Face Value'],
        'value':['₹'+list_[0].text+'cr','₹'+list_[1].text,'₹'+list_[2].text+'/'+'₹'+list_[3].text,list_[4].text,'₹'+list_[5].text,list_[6].text+'%',list_[7].text+'%',list_[8].text+'%','₹'+list_[9].text]}

    data_ = pd.DataFrame(data)
    print(data_)

    print(soup.find('div', class_ = 'pros').text.replace("\n"," "))

    print(soup.find('div', class_ = 'cons').text.replace('\n'," "))

    df2 = pd.read_html(f"https://www.screener.in/company/{i}/consolidated/")
    display(pd.concat([df2[0],df2[1]],axis = 1))
    display(pd.concat([df2[2],df2[3],df2[4],df2[5]],axis = 1))
    display(df2[6])
    display(df2[7])
    display(pd.concat([df2[8],df2[9]],axis = 1))
stock_market('url')
```
